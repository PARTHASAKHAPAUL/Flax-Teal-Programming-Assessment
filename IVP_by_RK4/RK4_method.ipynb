# Solving the given differential equation using the RK4 method:

    dy/dt = 1 - t^2 + y

```python
"""Given the differential equation dy/dt = 1 - t^2 + y"""
def f(t, y):
    return 1 - t**2 + y

"""Defining the RK4 method"""
def rk4(t_start, y0, t_end, n):
    """
    Parameters:
    t_start (float): Initial time
    t_end (float): End time
    n (int): Number of steps
    
    Initial Condition:
    y0 (float): Initial value of the solution at t_start
    
    Returns:
    list of float 't_values' : Times at which the solution was computed
    list of float 'y_values' : Values of the solution at the corresponding times
    """
    # Calculating step size
    h = (t_end - t_start) / n
    
    # Initializing two lists to store time and solution values
    t_values = [t_start]
    y_values = [y0]
    
    # Assigning initial values of t and y
    t = t_start
    y = y0
    
    # RK4 iterative method
    for _ in range(n):
        # Compute the intermediate values (k1, k2, k3, k4) using the RK4 method:
        # k1 is the slope at the beginning of the interval
        k1 = h * f(t, y)
        # k2 is the slope at the midpoint of the interval using k1
        k2 = h * f(t + 0.5 * h, y + 0.5 * k1)
        # k3 is another slope at the midpoint of the interval using k2
        k3 = h * f(t + 0.5 * h, y + 0.5 * k2)
        # k4 is the slope at the end of the interval using k3
        k4 = h * f(t + h, y + k3)

        # Update the solution y using the weighted average of the slopes
        y += (k1 + 2 * k2 + 2 * k3 + k4) / 6
        # Update the time variable t by the step size h
        t += h
        
        # Storing the computed value of y along with the corresponding time step 
        # in the corresponding lists that defined earlier 
        y_values.append(y)
        t_values.append(t)
    
    return t_values, y_values

# Given all the parameters as follows:
t_start = 0.0  # Initial value of t
t_end = 2.0  # Final value of t
n = 10  # Number of time steps
# Given the initial condition as follows:
y0 = 0.5 

# Solving the IVP by all the given conditions
t_vals, y_vals = rk4(t_start, y0, t_end, n)

# Printing the results y at given timesteps t
for t, y in zip(t_vals, y_vals):
    print(f"The solution of the given differential equation at t = {t:.2f}, is: y = {y:.4f}")
```

# Comparison of RK4 Numerical Solution with Exact Solution

```python
import matplotlib.pyplot as plt
import numpy as np

# Defining the Analytical solution found from the given differential equation
def exact_solution(t):
    return (t + 1)**2 - 0.5 * np.exp(t)

# Calculating the exact solution
t_fine = np.linspace(t_start, t_end, 30)  # A numpy array storing values from 0 to 2 with step size 30
y_exact = exact_solution(t_fine)          # Array of exact solutions of the previous t values

# Plotting
plt.figure(figsize=(8, 5))
# Plotting the RK4 Approximation with earlier t_vals and y_vals
plt.plot(t_vals, y_vals, 'o', label='RK4 Approximation')
# Plotting the Analytical Solution with previous t_fine and y_exact
plt.plot(t_fine, y_exact, '*', label='Exact Solution')
# Set the title of the plot
plt.title('Comparison of RK4 Numerical Solution with Exact Solution')
# Label the x-axis
plt.xlabel('Time (t)')
# Label the y-axis
plt.ylabel('Solution (y)')
# Enable the grid on the plot for better readability
plt.grid(True)
# Add a legend to the plot to describe plotted data (assuming legends have been defined elsewhere in the code)
plt.legend()
# Display the plot
plt.show()
```
